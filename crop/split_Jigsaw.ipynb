```python
def resize(height, width, num_columns, num_rows):
    
    size_col = num_columns*80
    size_row = num_rows*80

    resize_col = width / size_col 
    resize_row = height / size_row

    re_col = int(80 * resize_col)
    re_row = int(80 * resize_row)

    re_col_add = int(23 * resize_col)
    re_row_add = int(23 * resize_row)

    return resize_col ,resize_row, re_col, re_row, re_col_add, re_row_add
```

```python
def merge_image_grid(num_columns, num_rows, height ,width ,images ,resize_col ,resize_row):

    image_new = 255*np.ones((height *2, width *2, 3), dtype=np.uint8)   
    
    start_columns = 10 
    end_columns = 10
    end_rows = 10
    for j in range(num_columns): #3

        start_rows = 10
        for i in range(num_rows): #2
            
            random_index = np.random.randint(0, len(images))

            random_image_part = images.pop(random_index)

            end_columns =  random_image_part.shape[1] + start_columns 
            end_rows =   random_image_part.shape[0] + start_rows
            
            image_new[start_rows:end_rows ,start_columns:end_columns] = random_image_part

            start_rows = end_rows + int(resize_row * 46 )

        start_columns = end_columns + int(resize_col * 46)

    return image_new        
```

```python
images = split_image_jigsaw(num_columns, num_rows, height ,width ,image)
```

```python
image_new = merge_image_grid(num_columns, num_rows, height ,width ,images ,resize_col ,resize_row)
```
